## Splitting data

```python
X=df.drop(columns='diagnosis')
y=df['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=21,stratify=y)
```

## Encoding Target Variable

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y_train_enc=le.fit_transform(y_train)
```

## Feature Selection

```python
from sklearn.feature_selection import f_classif
f_scores, p_values = f_classif(X_train, y_train_enc)
result = pd.DataFrame({
    'Feature': X_train.columns.tolist(),
    'F_Score': f_scores,
    'P_Value': p_values
}).sort_values('F_Score', ascending=False)
```

```python
selected_features=result[result['P_Value'] < 0.05]['Feature'].tolist()
len(selected_features)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(result['Feature'], result['F_Score'], 
        color=['green' if p < 0.05 else 'red' for p in result['P_Value']])
plt.title('F-Scores (Green = Selected, Red = Dropped)')
plt.ylabel('F-Score')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Model training

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(class_weight='balanced')
model.fit(X_train[selected_features],y_train)
```

```python
y_pred=model.predict(X_test[selected_features])
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_pred,y_test)
accuracy
```

```python
from sklearn.ensemble import RandomForestClassifier
model_rf=RandomForestClassifier(class_weight='balanced')
model_rf.fit(X_train[selected_features],y_train)
```

```python
y_pred_rf=model_rf.predict(X_test[selected_features])
```

```python
accuracy_rf=accuracy_score(y_pred_rf,y_test)
accuracy_rf
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train[selected_features])  
X_test_scaled = scaler.transform(X_test[selected_features])        
model_lr = LogisticRegression(class_weight='balanced', max_iter=1000)
model_lr.fit(X_train_scaled, y_train)
```

```python
y_pred_lr=model_lr.predict(X_test_scaled)
accuracy_lr=accuracy_score(y_pred_lr,y_test)
accuracy_lr
```

```python
from sklearn.ensemble import BaggingClassifier
model_bc=BaggingClassifier(n_jobs=-1)
model_bc.fit(X_train[selected_features],y_train)
```

```python
y_pred_bc=model_bc.predict(X_test[selected_features])
accuracy_bc=accuracy_score(y_pred_bc,y_test)
accuracy_bc
```

Trained four different models and compared the accuracy.
Logistic Regression proved to be the best model
